### **Business Problem Understanding**

---

**Context**

Perusahaan properti di Daegu (daeguLiving) menyediakan layanan jual-beli apartemen melalui platform online. Salah satu tantangan yang dihadapi adalah menentukan harga apartemen yang kompetitif. Pemilik properti sering kali kesulitan menetapkan harga yang sesuai pasar karena tidak memiliki fitur yang memadai. Jika harga terlalu tinggi, apartemen sulit terjual. Sebaliknya, jika terlalu rendah, pemilik tidak mendapatkan keuntungan maksimal.

Daegu, melansir pada badan statistik resmi Korea Selatan, [KOSIS - Statistik Daegu ](https://kosis.kr/eng/) adalah kota terbesar keempat di Korea Selatan, memiliki populasi lebih dari 2,5 juta jiwa. Dikenal sebagai pusat industri, pendidikan, dan budaya, Daegu juga menawarkan berbagai fasilitas modern yang menarik bagi banyak pembeli apartemen. Salah satu daya tarik utama Daegu adalah keindahan alamnya yang dipadukan dengan perkembangan kota yang pesat. Beberapa kawasan, seperti  [Distrik Beomeo](https://www.mk.co.kr/en/special-edition/11008236) , adalah area yang sangat berkembang, sementara tempat seperti [Gunung Apsan](https://tour.daegu.go.kr/eng/index.do) menawarkan pemandangan yang indah, menarik banyak orang untuk membeli properti yang dekat dengan alam dan fasilitas kota.

Selain itu, Daegu adalah kota yang memiliki transportasi publik yang sangat terintegrasi dengan jaringan subway dan bus, dimana dapat diakses pada platform [DTRO](https://www.dtro.or.kr/eng/index.do). Jarak antara apartemen dan stasiun subway menjadi faktor penting yang mempengaruhi nilai properti, dimana sebagian besar kompleks apartemen dibangun di area ini. Maka dari itu, ukuran apartemen dan tipe apartemen biasa nya menjadi pertimbangan utama jika apartemen sudah berada daerah yang strategis. Memahami faktor-faktor ini akan membantu perusahaan dalam memberikan estimasi harga yang adil dan sesuai dengan kondisi pasar di Daegu.

---

**Problem Statement**

Pemilik apartemen kesulitan menetapkan harga jual yang kompetitif karena tidak ada fitur yang memprediksi harga apartemen berdasarkan faktor pasar dan karakteristik properti. Hal ini menyebabkan beberapa masalah:  
1. Apartemen dengan harga terlalu tinggi menjadi tidak laku sehingga memperpanjang waktu listing di platform.  
2. Apartemen dengan harga terlalu rendah justru merugikan pemilik properti.  
3. Kurangnya transparansi harga membuat calon pembeli sulit memperkirakan nilai apartemen.  

Jika masalah ini tidak diatasi, perusahaan bisa kehilangan peluang untuk menarik lebih banyak pemilik properti dalam menggunakan platform, memperlambat transaksi calon pembeli dan menurunkan daya saing platform.  

---

**Goals**

Berdasarkan permasalahan tersebut, perusahaan tentu perlu memiliki 'tool' yang dapat memprediksi serta membantu klien (dalam hal ini pemilik properti) untuk dapat menentukan harga apartement (Sale Price) secara akurat berdasarkan fitur properti. Terdapat perbedaan fitur pada masing-masing properti (apartemen), seperti tipe apartemen (ada atau tidaknya teras pribadi), akses transportasi umum, fasilitas terdekat area apartemen, sekolah terdekat, parking lot, tahun dibangunnya apartemen, fasilitas pribadi di apartemen dan luas apartemen.

Bagi perusahaan, prediction tool yang memberikan prediksi harga relevan, dapat meningkatkan pengalaman pengguna platform properti. Klien perusahaan yaitu Pemilik Properti akan mendapatkan harga estimasi yang kompetitif, kemudian calon pembeli properti juga dapat memiliki referensi harga yang jelas dan transparan. Hal ini tentunya akan menjadikan platform perusahaan unggul dalam daya saing dengan platform lain.

---

**Analytic Approach**

Jadi, yang perlu kita lakukan adalah menganalisis data untuk dapat menemukan pola dari fitur-fitur yang ada, yang membedakan harga apartemen satu dengan yang lainnya. 

Selanjutnya, kita akan membangun suatu model regresi yang akan membantu perusahaan untuk dapat menyediakan 'tool' prediksi harga jual apartemen yang baru masuk dalam daftar platform perusahaan, yang mana akan berguna untuk klien (Pemilik Properti) dalam menentukan harga jual apartemen nya.

**Metric Evaluation** 

Evaluasi metrik yang akan digunakan adalah RMSE, MAE & MAPE. Di mana RMSE adalah mengukur rata-rata akar kuadrat dari error prediksi. RMSE memberikan penalti lebih besar untuk error yang besar, cocok untuk memastikan akurasi estimasi harga. MAE adalah mengukur rata-rata error absolut antara prediksi dan nilai aktual. Cocok untuk memahami kesalahan rata-rata prediksi harga. Sedangkan MAPE adalah mengukur kesalahan rata-rata dalam bentuk persentase. Berguna untuk mengevaluasi seberapa jauh prediksi model dari nilai aktual dalam konteks bisnis.

Semakin kecil nilai RMSE, MAE, dan MAPE yang dihasilkan, berarti model semakin akurat dalam memprediksi harga sewa sesuai dengan limitasi fitur yang digunakan. 

- Dikarenakan tujuan kita adalah mengurangi kesalahan prediksi harga secara keseluruhan, maka RMSE menjadi metrics utama yang akan lebih diperhitungkan.

- Kemudian, metrics MAPE menjadi fokus kedua dikarenakan data kita memiliki harga yang sangat bervariasi atau ada kebutuhan untuk kesalahan relatif.

*Selain itu, kita juga bisa menggunakan nilai R-squared atau adj. R-squared jika model yang nanti terpilih sebagai final model adalah model linear. Nilai R-squared digunakan untuk menilai seberapa baik model menjelaskan varians data. Nilai mendekati 1 menunjukkan model sangat fit terhadap data. Namun, metrik ini tidak valid untuk model non-linear.*

```python
# Calculating Q1, Q3, and IQR
q1 = np.percentile(data_master['SalePrice'], 25)
q3 = np.percentile(data_master['SalePrice'], 75)
iqr = q3 - q1

# Calculating lower and upper bounds for outliers
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Identifying outliers
outliers = data_master[(data_master['SalePrice'] < lower_bound) | (data_master['SalePrice'] > upper_bound)]

# Visualizing using Boxplot
plt.figure(figsize=(10, 6))
plt.boxplot(data_master['SalePrice'], vert=False, patch_artist=True, boxprops=dict(facecolor="lightblue"))
plt.title("Boxplot of SalePrice", fontsize=14)
plt.xlabel("SalePrice", fontsize=12)
plt.show()
```

```python
# HallwayType (One Hot Encoding)
eda_df['HallwayType'] = eda_df['HallwayType'].map({'terraced': 0, 'mixed': 1, 'corridor': 2})

# TimeToSubway (Ordinal Encoding)
time_mapping = {'0-5min': 1, '5min~10min': 2, '10min~15min': 3, '15min~20min': 4, 'no_bus_stop_nearby': 5}
eda_df['TimeToSubway'] = eda_df['TimeToSubway'].map(time_mapping)

# SubwayStation (Binary Encoding)
subway_mapping = {'Kyungbuk_uni_hospital': 0, 'Chil-sung-market': 1, 'Bangoge': 2, 'Sin-nam': 3, 
                  'Banwoldang': 4, 'no_subway_nearby': 5, 'Myung-duk': 6, 'Daegu': 7}
eda_df['SubwayStation'] = eda_df['SubwayStation'].map(subway_mapping)

correlation_matrix = eda_df.corr()

# Plotting korelasi dengan heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Features')
plt.show()
```

### **Data Preprocessing**

Dalam mengubah tipe data kolom menjadi numerik, ada 3 pilihan Encoding:
- One Hot Encoding
    - Hanya berlaku di kolom yang Category nya nominal atau level nya sama
    - Efektif jika 5 data saja yang di Encoding agar menghindari Overfitting
    - Jika lebih dari 5 data, tidak efektif karena One Hot Encoding akan menghasilkan kolom baru sesuai degan banyak nya unique data di kolom

- Binary Encoding
    - Sama seperti One Hot Encoding (data yang nominal atau level nya sama) namun untuk data yang lebih dari 5
    - Cara kerjanya juga sama, akan menghasilkan kolom baru tapi satu kolom saja dengan langsung encoding nya (tidak satuan angka)

- Ordinal Encoding
    - Digunakan untuk data category yang berurutan/bertingkat. Contoh: SD, SMP, SMA
    - Langsung Encoding data tanpa bikin kolom baru

Sebagai catatan, kolom `HallwayType` dan `SubwayStation` akan dikecualikan dari scaling karena ini adalah kategori yang terpisah dimana nilainya terbatas hanya 0 dan 1.
```python
# Definisikan transformer
transformer = ColumnTransformer([
    # Untuk One Hot Encoding pada 'HallwayType'
    ('OneHotEncoder', OneHotEncoder(), ['HallwayType']),
    
    # Untuk Ordinal Encoding pada 'TimeToSubway'
    ('OrdinalEncoder', OrdinalEncoder(categories=[['0-5min', '5min~10min', '10min~15min', '15min~20min', 'no_bus_stop_nearby']]), ['TimeToSubway']),
    
    # Untuk Binary Encoding pada 'SubwayStation'
    ('BinaryEncoder', BinaryEncoder(), ['SubwayStation']),
    
    # Untuk Scaling kolom numerik lainnya setelah log transform
    ('RobustScaler', RobustScaler(),  ['YearBuilt', 'N_FacilitiesInApt','Total_Nearby_Facilities' ,'Size(sqf)_log'])
], remainder='passthrough')
```

```python
# Bikin list untuk store results
listMAEmean = []
listMAEstd = []
listMAPEmean = []
listMAPEstd = []
listMSEmean = []
listMSEstd = []
listRMSEmean = []
listRMSEstd = []
# Looping over each model for cross-validation
for model in normalModel:
    pipeModel = Pipeline([('prepro', transformer), ('modeling', model)])
    cvScore = cross_validate(estimator=pipeModel,
                             X=X_train,
                             y=y_train,
                             cv=KFold(n_splits=5, shuffle=True, random_state=302),
                             scoring=['neg_mean_absolute_error',
                                      'neg_mean_absolute_percentage_error',
                                      'neg_mean_squared_error',
                                      'neg_root_mean_squared_error',
                                      'r2'])
    
    # cross validation
    listMAEmean.append(-cvScore['test_neg_mean_absolute_error'].mean())
    listMAEstd.append(cvScore['test_neg_mean_absolute_error'].std())
    
    listMAPEmean.append(-cvScore['test_neg_mean_absolute_percentage_error'].mean())
    listMAPEstd.append(cvScore['test_neg_mean_absolute_percentage_error'].std())
    
    listMSEmean.append(-cvScore['test_neg_mean_squared_error'].mean())
    listMSEstd.append(cvScore['test_neg_mean_squared_error'].std())
    
    listRMSEmean.append(-cvScore['test_neg_root_mean_squared_error'].mean())
    listRMSEstd.append(cvScore['test_neg_root_mean_squared_error'].std())

# Buat dataframe
dfCV = pd.DataFrame()
dfCV['algo'] = normalModelName
dfCV['MAEmean'] = listMAEmean
dfCV['MAEstd'] = listMAEstd
dfCV['MAPEmean'] = listMAPEmean
dfCV['MAPEstd'] = listMAPEstd
dfCV['MSEmean'] = listMSEmean
dfCV['MSEstd'] = listMSEstd
dfCV['RMSEmean'] = listRMSEmean
dfCV['RMSEstd'] = listRMSEstd

# Display
dfCV
```

#### Cek Overfitting/Ideal/Underfitting

Disini kita akan memasikan bahwa model harus 'Ideal' dahulu sebelum proses Hyperparameter Tuning. Hal ini bertujuan agar proses tuning dapat menggeneralisasi dan tidak terlalu mempelajari noise atau pola yang ada.

Maka dari itu kita akan melakukan perbandingan performa prediksi pada training set dan test set sebelumnya. Jika perbedaan antara keduanya besar, berarti model bisa overfitting atau underfitting.

```python
# Evaluasi Overfitting/Ideal/Underfitting
models = {
    'svreg' : SVR(),
    'gdreg' : GradientBoostingRegressor(random_state=302)
}

# List untuk menyimpan hasil
results = []

for name, model in models.items():
    pipeline = Pipeline([
        ('preprocessing', transformer),
        ('model', model)
    ])

    # Train model
    pipeline.fit(X_train, y_train)

    # Evaluasi pada training set
    y_train_pred = pipeline.predict(X_train)
    y_train_pred = np.expm1(y_train_pred)  # Balikkan log pada prediksi

    # Evaluasi pada test set
    y_test_pred = pipeline.predict(X_test)
    y_test_pred = np.expm1(y_test_pred)  # Balikkan log pada prediksi

    # Simpan hasil
    results.append({
        'Model': name,
        'Train RMSE': np.sqrt(mean_squared_error(np.expm1(y_train), y_train_pred)),
        'Test RMSE': np.sqrt(mean_squared_error(np.expm1(y_test), y_test_pred)),
        'Train MAE': mean_absolute_error(np.expm1(y_train), y_train_pred),
        'Test MAE': mean_absolute_error(np.expm1(y_test), y_test_pred),
        'Train MAPE': mean_absolute_percentage_error(np.expm1(y_train), y_train_pred),
        'Test MAPE': mean_absolute_percentage_error(np.expm1(y_test), y_test_pred)
    })

# Tampilkan hasil sebagai DataFrame
df_results = pd.DataFrame(results)
df_results
```

Evaluasi:

#### Hyperparameter Tuning

Di tahap ini kita akan meningkatkan performa model GradientBoostingRegressor yang sudah dipilih. Untuk panduan tuning model dapat merujuk ke [Tuning GradientBoostingRegressor](https://scikit-learn.org/1.5/modules/generated/sklearn.ensemble.GradientBoostingRegressor.html).

```python
# Pipeline dengan preprocessing dan model
pipeline = Pipeline([
    ('preprocessing', transformer),
    ('model', GradientBoostingRegressor(random_state=302))
])

# Parameter grid for hyperparameter tuning (with 'model__' prefix for parameters inside the pipeline)
param_grid = {
    'model__n_estimators': [50, 100, 150, 200],  # Jumlah estimators (pohon keputusan)
    'model__learning_rate': [0.01, 0.05, 0.1, 0.2],  # Learning rate (kecepatan belajar)
    'model__max_depth': [3, 4, 5, 6],  # Kedalaman pohon keputusan
    'model__min_samples_split': [2, 5, 10],  # Jumlah minimum sampel untuk memecah simpul
    'model__min_samples_leaf': [1, 2, 4],  # Jumlah minimum sampel di daun
    'model__subsample': [0.8, 0.9, 1.0],  # Subsample untuk setiap estimasi
    'model__max_features': ['sqrt', 'log2', None]  # Fitur maksimum untuk membagi simpul
}
```

Disini kita menggunakan metrics RMSE untuk refitting model yang telah di tuning. Alasan fokus ke RMSE karena metrics ini memberikan nilai yang menunjukkan seberapa besar rata-rata selisih harga prediksi apartemen dengan harga actual apartemen. Metrics ini cocok untuk memastikan akurasi estimasi harga sesuai dengan konteks masalah kita.

```python
# Setup RandomSearchCV
random_search = RandomizedSearchCV(
    estimator=pipeline,
    param_distributions=param_grid,
    scoring={
        'rmse': 'neg_root_mean_squared_error',
        'mae': 'neg_mean_absolute_error',
        'mape': 'neg_mean_absolute_percentage_error'
    },
    cv=KFold(n_splits=10, shuffle=True, random_state=302),
    n_iter=100,
    refit='rmse',  # Refitting based on RMSE (choose the metric you want to optimize for)
    n_jobs=-1
)

# Fitting RandomSearchCV
random_search.fit(X_train, y_train)
```

```python
# Menampilkan hasil terbaik
print("Best Parameters: ", random_search.best_params_)
print("Best RMSE (Negative): ", random_search.best_score_)
```

Parameter pada GradientBoostingRegressior agar menghasilkan performa terbaik berdasarkan metrik utama **neg_root_mean_squared_error (RMSE)**, yaitu:

- model__subsample = 0.9
- model__n_estimators = 200
- model__min_samples_split = 2
- model__min_samples_leaf = 1
- model__max_features = log2
- model__max_depth = 3
- model__learning_rate = 0.1
    
Kemudian skor terbaik dari neg_root_mean_squared_error (RMSE) menunjukkan bahwa rata-rata akar kesalahan kuadrat prediksi model pada validasi silang adalah sekitar -0.223. Nilai ini menunjukkan performa yang baik, mengingat skala target log-transformasi.

Nah, setelah mendapatkan parameter terbaik untuk model GradientBoostingRegressior, kita akan mengevaluasi parameter tersebut ke semua metrics agar memastikan bahwa parameter hasil tuning juga memiliki performa baik ke metrics lain (tidak hanya RSME).

```python
# Menampilkan hasil dari semua metrik
results = random_search.cv_results_
for metric in ['mean_test_rmse', 'mean_test_mae', 'mean_test_mape']:
    best_index = random_search.best_index_
    print(f"Best {metric}: {results[metric][best_index]}")

# Skor terbaik
best_rmse = random_search.best_score_
```

- RMSE: Hasil yang sama seperti sebelumnya, menunjukkan bahwa akar rata-rata kesalahan kuadrat terbaik adalah -0.223.
- MAE (Mean Absolute Error): Kesalahan absolut rata-rata terbaik adalah -0.181, yang memberikan gambaran alternatif terhadap kesalahan prediksi model.
- MAPE (Mean Absolute Percentage Error): Persentase kesalahan terbaik adalah sekitar 1.5%, menunjukkan bahwa prediksi model sangat dekat dengan nilai sebenarnya dalam persentase.


Sesuai hasil diatas, dapat kita validasi bahwa **parameter hasil Hyperparameter Tuning sudah optimal untuk model GradientBoostingRegressor**. Dikarenakan model menunjukkan performa yang konsisten dengan metrics selain RMSE (MAE dan MAPE), dimana hasil nya yang rendah pada validasi silang.
